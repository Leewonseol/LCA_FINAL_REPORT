# 실습과제: 병원 환자 데이터베이스 관리

병원에서 제공하는 기본 환자 데이터베이스를 활용하여 환자 정보를 조회, 수정, 추가, 삭제하는 프로그램을 작성해보세요.

### 요구사항

1. **환자 정보 조회**
   - 첫 번째 환자의 모든 정보 출력 (get() 메서드 활용)
   - 딕셔너리 키별로 안전한 데이터 접근

2. **환자 정보 수정**
   - 두 번째 환자의 질병 정보 수정 (update() 메서드 활용)
   - 수정 전후 상태 비교 출력

3. **새 환자 등록**
   - 새로운 환자 데이터를 딕셔너리로 생성 후 데이터베이스에 추가
   - 추가된 환자 정보를 데이터베이스에서 다시 조회하여 출력

4. **환자 퇴원 처리**
   - 마지막 환자를 데이터베이스에서 제거
   - 퇴원한 환자의 정보 출력

5. **통계 정보**
   - 환자들의 나이 통계 (최고령, 최연소, 평균)
   - 내장 함수 활용 (max, min, sum)

6. **출력 형식**
```
=== 첫 번째 환자 정보 ===
환자번호: P001
이름: 김환자
나이: 65세
진료과: 내과
질병: 고혈압

=== 환자 질병 수정 전 ===

이름: 이환자
기존 질병: 맹장염

=== 환자 질병 수정 후 ===
환자번호: P002
이름: 이환자
수정된 질병: 급성맹장염

=== 새 환자 등록 ===
환자번호: P004
이름: 홍환자
나이: 40세
진료과: 피부과
질병: 아토피
현재 총 환자 수: 4명

=== 환자 퇴원 처리 ===
퇴원 환자번호: P004
퇴원 환자명: 홍환자
퇴원 후 총 환자 수: 3명

=== 환자 나이 통계 ===
최고령 환자 나이: 65세
최연소 환자 나이: 32세
평균 나이: 47.3세

==================================================
최종 병원 환자 데이터베이스 현황
==================================================
총 입원 환자: 3명
```

````python

# @title 정답

# 기본 데이터 - 병원 환자 데이터베이스
patients_db = {
    "P001": {"name": "김환자", "age": 65, "department": "내과", "disease": "고혈압"},
    "P002": {"name": "이환자", "age": 45, "department": "외과", "disease": "맹장염"},
    "P003": {"name": "박환자", "age": 32, "department": "소아과", "disease": "감기"}
}

# 1. 특정 환자 정보 조회 (첫 번째 환자) - get() 메서드 활용
# 딕셔너리 형태이므로 인덱스 접근이 아닌 키 접근으로 변경해야 합니다.
first_patient_id = "P001"
first_patient = patients_db[first_patient_id]
print("=== 첫 번째 환자 정보 ===")
print(f"환자번호: {first_patient_id}")
print(f"이름: {first_patient.get('name', '정보없음')}")
print(f"나이: {first_patient.get('age', 0)}세")
print(f"진료과: {first_patient.get('department', '정보없음')}")
print(f"질병: {first_patient.get('disease', '정보없음')}")

# 2. 특정 환자 질병 수정 (두 번째 환자) - update() 메서드 활용
# 딕셔너리 형태이므로 인덱스 접근이 아닌 키 접근으로 변경해야 합니다.
second_patient_id = "P002"
second_patient = patients_db[second_patient_id]

print(f"\n=== 환자 질병 수정 전 ===")
print(f"환자번호: {second_patient_id}")
print(f"이름: {second_patient.get('name', '정보없음')}")
print(f"기존 질병: {second_patient.get('disease', '정보없음')}")

second_patient.update({"disease": "급성맹장염"})  # update() 메서드로 질병 수정

print(f"\n=== 환자 질병 수정 후 ===")
print(f"환자번호: {second_patient_id}")
print(f"이름: {second_patient.get('name', '정보없음')}")
print(f"수정된 질병: {second_patient.get('disease', '정보없음')}")

# 3. 새로운 환자 추가
new_patient_id = "P004"
new_patient_data = {"name": "홍환자", "age": 40, "department": "피부과", "disease": "아토피"}
patients_db[new_patient_id] = new_patient_data # 딕셔너리에 새 키-값 추가

# 새로 추가된 환자 정보를 데이터베이스에서 조회
latest_patient = patients_db.get(new_patient_id) # 새로 추가된 환자 ID로 접근
print(f"\n=== 새 환자 등록 ===")
print(f"환자번호: {new_patient_id}")
print(f"이름: {latest_patient.get('name', '정보없음')}")
print(f"나이: {latest_patient.get('age', 0)}세")
print(f"진료과: {latest_patient.get('department', '정보없음')}")
print(f"질병: {latest_patient.get('disease', '정보없음')}")
print(f"현재 총 환자 수: {len(patients_db)}명")

# 4. 환자 퇴원 처리 (마지막 환자 제거)
# 딕셔너리는 순서가 없으므로 '마지막 환자'라는 개념이 없습니다.
# 여기서는 'P004'를 퇴원 처리하는 것으로 가정하겠습니다.
discharged_patient_id = new_patient_id
discharged_patient = patients_db.pop(discharged_patient_id) # 특정 키를 이용해 제거

print(f"\n=== 환자 퇴원 처리 ===")
print(f"퇴원 환자번호: {discharged_patient_id}")
print(f"퇴원 환자명: {discharged_patient.get('name', '정보없음')}")
print(f"퇴원 후 총 환자 수: {len(patients_db)}명")

# 5. 통계 함수 활용 (반복문/조건문 없이)
# 딕셔너리의 모든 값(환자 정보 딕셔너리)에서 'age'를 추출합니다.
ages = [patient_data.get("age", 0) for patient_data in patients_db.values()]

print(f"\n=== 환자 나이 통계 ===")
print(f"최고령 환자 나이: {max(ages)}세")
print(f"최연소 환자 나이: {min(ages)}세")
print(f"평균 나이: {sum(ages) / len(ages):.1f}세")

# 6. 최종 데이터베이스 현황
print(f"\n{'='*50}")
print("최종 병원 환자 데이터베이스 현황")
print(f"{'='*50}")
print(f"총 입원 환자: {len(patients_db)}명")
````
